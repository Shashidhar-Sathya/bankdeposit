```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt
import joblib  # For loading saved models
import pandas as pd
```

```python
files = pd.read_csv('savedmodels/list.csv')
savedmodels = files['filename'].apply(lambda x: f'savedmodels/{x}').tolist()
savedmodels
```

```python
def plot_roc_auc(model_filenames):

    # Load the test data 
    X_test = pd.read_csv('data/X_test_resampled.csv', index_col=0 )
    y_test = pd.read_csv('data/y_test_resampled.csv', index_col=0 )

    plt.figure(figsize=(6, 4))

    # Load the saved models and their labels
    for filename, label in model_filenames:
        # Load the saved model
        loaded_model = joblib.load(filename) 

        # Get probabilities for positive class
        y_pred_proba = loaded_model.predict_proba(X_test)[:, 1]  
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{label} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curves')
    plt.legend(bbox_to_anchor=(1.05, 0.5), loc='upper left')
    plt.show()
```

```python
model_filenames = [('savedmodels/01_baseline_model.pkl', 'Baseline Model'),
                   ('savedmodels/02_logistic_model.pkl', 'Simple Logistic Regression')]
plot_roc_auc(model_filenames=model_filenames)
```

```python
model_filenames = [('savedmodels/03_pipe_decisiontreeclassifier_defaults.pkl','Decision Tree-Defaults'),
                   ('savedmodels/03_pipe_knn_defaults.pkl','KNN Classifier-Defaults'),
#                    ('savedmodels/03_pipe_svc_defaults.pkl','SVC Classifier-Defaults'),
                   ('savedmodels/03_pipe_randomforest_defaults.pkl','RandomForest Classifier-Defaults')]
plot_roc_auc(model_filenames=model_filenames)
```

```python
def plot_roc_auc_pre_rec(model_info):
    # Load the test data 
    X_test = pd.read_csv('data/X_test_resampled.csv', index_col=0 )
    y_test = pd.read_csv('data/y_test_resampled.csv', index_col=0 )

    plt.figure(figsize=(10,4))

    # Plot ROC curves
    plt.subplot(1, 2, 1)  # Create a subplot for ROC curves
    for filename, label in model_info:

        # Load saved models      
        loaded_model = joblib.load(filename)
        y_pred_proba = loaded_model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{label} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curves')
    plt.legend(bbox_to_anchor=(0.4,-0.45), loc='lower center', borderaxespad=0.)

    # Plot Precision-Recall curves
    # Create a subplot for Precision-Recall curves
    plt.subplot(1, 2, 2)  
    for filename, label in model_info:
        loaded_model = joblib.load(filename)
        y_pred_proba = loaded_model.predict_proba(X_test)[:, 1]
        precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
        
        # Calculate AUC for Precision-Recall curve
        pr_auc = auc(recall, precision)  
        plt.plot(recall, precision, lw=2, label=f'{label} (AUC = {pr_auc:.2f})')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curves')
#     plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.legend(bbox_to_anchor=(0.4, -0.45), loc='lower center', borderaxespad=0.)
    plt.tight_layout()  # Adjust spacing between subplots
    plt.show()
```

```python
model_filenames = [('savedmodels/01_baseline_model.pkl', 'Baseline Model'),
                   ('savedmodels/02_logistic_model.pkl', 'Simple Logistic Regression')
                  ]
plot_roc_auc_pre_rec(model_info=model_filenames)
```

```python
model_filenames = [('savedmodels/03_pipe_decisiontreeclassifier_defaults.pkl','Decision Tree-Defaults'),
                   ('savedmodels/03_pipe_knn_defaults.pkl','KNN Classifier-Defaults'),
#                    ('savedmodels/03_pipe_svc_defaults.pkl','SVC Classifier-Defaults'),
                   ('savedmodels/03_pipe_randomforest_defaults.pkl','RandomForest Classifier-Defaults')]
plot_roc_auc_pre_rec(model_info=model_filenames)
```
